```python
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams.update({'font.size':15})

T = 1e10*np.linspace(0.01,6,1000) #Temperature
k = 8.6e-5 #Boltzmann constant in eV/K
nn_np = np.exp(-1.29e6/(k*T))
plt.plot(T, nn_np)
plt.axis([6e10,0, 0, 1])
plt.xlabel('Temperature/K')
plt.ylabel('$n_n/n_p$')
plt.show()
```

```python
T = 1e9*np.linspace(0.1,1,1000) #Temperature
BD = 2.22e6
mnc2 = 939.6e6
eta = 6.1e-10
nD_nn = 6.5*eta*((k*T/mnc2)**(3/2))*np.exp(BD/(k*T))
plt.plot(T, nD_nn)
plt.axis([1e9,6e8, 0.001,10000])
plt.yscale('log')
plt.xlabel('Temperature/K')
plt.ylabel('$n_D/n_n$')
plt.show()
```
